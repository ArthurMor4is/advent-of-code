```python
input = '''....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#...'''
```

```python
input_matrix = [list(inp) for inp in input.split('\n')]
```

```python
M, N = len(input_matrix), len(input_matrix[0])
```

```python
direction = {
    '^': (-1, 0),
    '>': (0, 1),
    'v': (1, 0),
    '<': (0, -1)
}

rotated = {
    '^': '>',
    '>': 'v',
    'v': '<',
    '<': '^'
}
```

```python
stop = False
for row in range(M):
    if stop:
        break
    for col in range(N):
        guard_position = [row, col]
        symbol = input_matrix[row][col]
        if symbol in ['^', '>', 'v', '<']:
            stop = True
            break

def next_step():
    global guard_position
    global symbol

    cur_row, cur_col = guard_position
    if cur_row == M - 1 or cur_col == N - 1:
        return
    dx, dy = direction[symbol]
    new_row, new_col = cur_row + dx, cur_col + dy
    
    if input_matrix[new_row][new_col] == "#":
        symbol = rotated[symbol]
        input_matrix[cur_row][cur_col] = symbol
        return
    
    input_matrix[cur_row][cur_col] = '.'
    input_matrix[new_row][new_col] = symbol
    guard_position = [new_row, new_col]
    return

visited = set()
visited.add(tuple(guard_position))
while 0 < guard_position[0] < M - 1 and 0 < guard_position[1] < N - 1:
    next_step()
    visited.add(tuple(guard_position))
```
